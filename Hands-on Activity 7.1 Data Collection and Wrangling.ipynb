```python
import pandas as pd

fb = pd.read_csv("fb.csv")
ap =  pd.read_csv("aapl.csv")
am =  pd.read_csv("amzn.csv")
nfx =  pd.read_csv("nflx.csv")
gg =  pd.read_csv("goog.csv")
```

```python
# adding the ticker

fb["ticker"] = "FB"
ap["ticker"] = "AAPL"
am["ticker"] = "AMZN"
nfx["ticker"] = "NFLX"
gg["ticker"] = "GOOG"
```

```python
# appending to a single data frame
faang = pd.concat([fb, ap, am, nfx, gg])
```

```python
# saving as faang
faang.to_csv("faang.csv", index=False)
```

## Exercise 3


```python
import requests
from io import StringIO

response = requests.get(
    "https://raw.githubusercontent.com/mGalarnyk/datasciencecoursera/refs/heads/master/2_R_Programming/data/hospital-data.csv", 
    params={"$limit":50_000}
)
if response.ok:
    payload = pd.read_csv(StringIO(response.text))
else:
    print(f"Request was not successful and returned code: {response.status_code}.")
    payload = None
```

```python
payload
```

```python
hospital = payload[["Hospital Name","Address 1","Phone Number"]]
```

```python
hospital.to_csv("hospital.csv", index=False)
```
